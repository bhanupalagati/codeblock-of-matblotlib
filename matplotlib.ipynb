# Example 1 :- Plotting sine-angle graph
1. create various angles
2. calculate sin value for those angles
3. plot the graph
4. Add titles and other things
5. Show the Graph

```python
import matplotlib.pyplot as plt
import numpy as np
import math
# Create angle and sin values
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('Sine-Radian Graph')
plt.show()
```

# Using Axes in the Matplotlib
Although pyplot is easy and handy. We sometimes need more control over the plot.
For example if a canvas has multiple plots.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()# this will create an empty canvas
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('angle')
ax.set_ylabel('sine')
ax.set_title('Sine-Radian Graph')
plt.show()
```

# Axes with multiple plots
This pictures sin, tan, and cos

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, .1)
sinx = np.sin(x)
cosecx = np.arccos(x)
cosx = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('The Three curves')
ax.set_xlabel('Radians')
ax.set_ylabel('projected values of sin cos and tan')
ax.plot(x, sinx, 'go-')
ax.plot(x, cosecx, 'bH-.')
ax.plot(x, cosx, 'r+--')
plt.show()
```

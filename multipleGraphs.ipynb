# Here we will see about various ways to create multiple graphs on one canvas
1. We can use matplotlib add_subplot(nrows, ncols, index) fn
2. Or we can create multiple axes and plot with those axes on the created fig

First let's see the subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.1)
y = np.sin(x)
cosx = np.cos(x)

# Let's start with the plotting
fig = plt.figure()
axes2 = fig.add_subplot(1, 1, 1, facecolor='y')
axes3 = fig.add_subplot(2, 2, 4, facecolor='k')
axes2.plot(x, x, 'b-.')
axes3.plot(x, cosx, 'g+')
plt.show()
```

# Let's see the process with multiple axes instead of subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.1)
y = np.sin(x)
cosx = np.cos(x)

# Let's start with the plotting
fig = plt.figure()
axes2 = fig.add_axes([0,0,1,1])
axes3 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
axes2.plot(x, y, 'b-.')
axes3.plot(x, cosx, 'g+')
plt.show()
```
